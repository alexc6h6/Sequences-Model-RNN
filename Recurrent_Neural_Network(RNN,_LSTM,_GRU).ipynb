# Recurrent Neural Network

Recurrent neural network(RNN), is one of the deep neural networks model that used for sequence tasks, such as machine translation, question answering, text sentiment classification, and music generation.

Recurrent neural network pass the cell state(a) as memory that are stored and input to the neuron for the next time step. The neural remembers the previous inputs(x) and predicts the output(y) by following the previous inputs. The neural networks share the same parameters across inputs and updated the parameters for each batch dataset.

```
data = pd.read_csv('/content/drive/My Drive/imdb/IMDB Dataset.csv')
```

```
print(data.head())
```

```
x_train, x_val, y_train, y_val = train_test_split(data['review'], data['sentiment'], test_size = 0.1)

print(x_train.shape)
```

```
le = LabelEncoder()
y_train = le.fit_transform(y_train).reshape(-1, 1)
y_val = le.fit_transform(y_val).reshape(-1, 1)

print(y_train.shape)
```

```
max_words = 20000
max_length = 200
```

```
token = Tokenizer(num_words = max_words, oov_token = '<oov>')
token.fit_on_texts(x_train)

print(token.word_index)
print(len(token.word_index))
```

```
#transfrom the words to numerical token in the sequences

train_sequences = token.texts_to_sequences(x_train)
val_sequences = token.texts_to_sequences(x_val)

train_padded = pad_sequences(train_sequences, maxlen = max_length, padding = 'post')
val_padded = pad_sequences(val_sequences, maxlen = max_length, padding = 'post')

print(train_padded.shape)
```

```
reduce_lr = ReduceLROnPlateau(monitor = 'val_loss', verbose = 1, min_lr = 0.000001, patience = 5)
```

```
RNN_model = tf.keras.models.Sequential([Embedding(input_dim = max_words, output_dim = 32, input_length = max_length),
                                    Bidirectional(SimpleRNN(64, return_sequences = True)),
                                    Bidirectional(SimpleRNN(64)),
                                    Dropout(0.25),
                                    Dense(64, kernel_regularizer = regularizers.l2(0.1)),
                                    BatchNormalization(),
                                    Activation('relu'),
                                    Dropout(0.25),
                                    Dense(1, activation = 'sigmoid')])

RNN_model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])

RNN_model.summary()
plot_model(RNN_model, show_shapes = True)
```

```
batch = 128
epochs = 20


history = RNN_model.fit(train_padded, y_train,
                    batch_size = batch,
                    epochs = epochs,
                    verbose = 1, 
                    callbacks = [reduce_lr],
                    validation_data = (val_padded, y_val))
```

```

def show_train_history_acc(history):
    plt.plot(history.history['acc'])
    plt.plot(history.history['val_acc'])
    plt.title('Train History')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()

def show_train_history_loss(history):
    plt.plot(history.history["loss"])
    plt.plot(history.history["val_loss"])
    plt.title("Train History")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()

show_train_history_acc(history)
show_train_history_loss(history)
```

```
from tensorflow.keras import regularizers

model_LSTM = tf.keras.models.Sequential([tf.keras.layers.Embedding(max_words, 32, input_length = max_length),
                                    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64, return_sequences = True)),
                                    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
                                    tf.keras.layers.Dropout(0.5),
                                    tf.keras.layers.Dense(64, kernel_regularizer = regularizers.l2(0.1)),
                                    tf.keras.layers.BatchNormalization(),
                                    tf.keras.layers.Activation('relu'),
                                    tf.keras.layers.Dropout(0.5),
                                    tf.keras.layers.Dense(1, activation = 'sigmoid')])

model_LSTM.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])

model_LSTM.summary()
```

```
history = model_LSTM.fit(train_padded, y_train, 
                         batch_size = 128, 
                         epochs = 10, 
                         verbose= 1,
                         validation_data = (val_padded, y_val))
```

```

show_train_history_acc(history)
show_train_history_loss(history)
```
